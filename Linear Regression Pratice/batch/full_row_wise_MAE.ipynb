```python
import numpy as np
from numpy import genfromtxt
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data= genfromtxt('../data/advertising.csv', delimiter= ',', skip_header= 1)
```

```python
X= data[:, :3]
y= data[:, 3:]
```

```python
X_b= np.hstack([np.ones((X.shape[0], 1)), X])
X_b.shape
```

```python
theta = np.random.randn(4, 1)
```

```python
n_epochs = 50
epsilons = 1e-10
lr = 0.0005
losses = []

for epoch in range(n_epochs):
    y_hat = X_b @ theta                           # (n,1)
    loss = np.sum(np.abs(y - y_hat)) / len(y)     # scalar
    print(f"loss: {loss}")
    losses.append(loss)

    # k (n,1)
    k = (y - y_hat) / (np.abs(y - y_hat) + epsilons)

    # gradient (n_features+1,1)
    gradient = -(X_b.T @ k) / len(y)

    # update
    theta = theta - lr * gradient

plt.plot(losses)
```
