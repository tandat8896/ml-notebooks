```python
import numpy as np
data = np.genfromtxt('../M05W02_Fri/advertising.csv' , delimiter=',', skip_header=1)
X= data[:, :3]
y= data[:, 3:]
X_T = X.T
theta = np.random.randn(4,1)
lr=0.00001
n_features, n_samples = X_T.shape
X_b = np.vstack([np.ones((1, n_samples)), X_T])
n_epochs = 50
losses = []
for epoch in range(n_epochs):
    
    y_hat = theta.T@X_b
    loss = (1/n_samples)*(y_hat - y.T).dot((y_hat - y.T).T)
    losses.append(loss[0][0])
    k = 2*(y_hat - y.T)
    gradient = np.multiply(np.vstack((k,k,k,k,)), X_b)
    gradient = gradient.dot(np.ones((n_samples, 1))) / n_samples    
    theta = theta - lr*gradient

    
```
