```python
import torch
import torch.nn.functional as F

# Inputs
X = torch.tensor([[-0.1, 0.1, 0.3],
                  [0.4, -1.1, -0.3]])
C = torch.tensor([[-0.6 ,0.3, -0.4],
                  [0.5, 0.9, -0.5]])

# Weights
W_Q = torch.tensor([[-0.35, 0.51, 0.50],
                    [0.36, -0.47, -0.29],
                    [-0.51, -0.14, -0.56]])
W_K = torch.tensor([[-0.49, -0.68, 0.18],
                    [-0.44, -0.46, 0.18],
                    [0.07, -0.10, 0.44]])
W_V = torch.tensor([[-0.41, 0.39, -0.65],
                    [-0.40, -0.07, -0.34],
                    [-0.55, -0.13, -0.29]])
W_O = torch.tensor([[-0.36, -0.08, 0.32],
                    [0.27, 0.05, 0.15],
                    [-0.05, -0.28, 0.05]])

d = 3.0

# Case 1: Standard Cross Attention (Q from X, K/V from C)
Q = X @ W_Q
K = C @ W_K
V = C @ W_V
scores = (Q @ K.T) / (d**0.5)
attn = F.softmax(scores, dim=-1)
context = attn @ V
Y_standard = context @ W_O

# Case 2: Swapped Cross Attention (Q from C, K/V from X)
Q2 = C @ W_Q
K2 = X @ W_K
V2 = X @ W_V
scores2 = (Q2 @ K2.T) / (d**0.5)
attn2 = F.softmax(scores2, dim=-1)
context2 = attn2 @ V2
Y_swapped = context2 @ W_O

print("Standard Result (Q=X, K=C):")
print(Y_standard)
print("Rounded Standard:")
print(torch.round(Y_standard * 100) / 100)

print("\nSwapped Result (Q=C, K=X):")
print(Y_swapped)
```
